```python

def points_to_grid(points, grid_shape, resolution):
    """
    Convert a list of points to a binary grid.
    
    :param points: List of points in N-dimensional space
    :param grid_shape: Shape of the grid (number of cells in each dimension)
    :param resolution: Size of each cell
    :return: N-dimensional binary grid
    """
    grid = np.zeros(grid_shape, dtype=int)
    for point in points:
        # Calculate the grid cell index for the point
        indices = tuple((np.array(point) / resolution).astype(int))
        # Set the corresponding cell to 1
        grid[indices] = 1
    return grid
```
